```python
import pandas as pd
df = pd.read_csv("C:/Users/Admin/Desktop/Online Retail.csv", encoding='ISO-8859-1')
df.head(20)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df["Revenue"]=df["UnitPrice"]*df['Quantity']
df
```

```python
grouped_data = df.groupby('CustomerID').agg({
    'Revenue': 'sum',
}).reset_index()
grouped_data=grouped_data.sort_values(by="Revenue",ascending=False).reset_index(drop=True)
```

```python
revenue=grouped_data.head(10)
revenue
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,5))
CustomerID=[14646.0,18102.0,17450.0,14911.0,12415.0,14156.0,17511.0,16684.0,13694.0,15311.0,]
p=np.arange(len(CustomerID))
width=0.35
plt.bar(p,revenue["Revenue"],width,color="blue",label="Revenue")
plt.xticks(p,CustomerID)
plt.xlabel("CustumerID")
plt.ylabel("Revenue")
plt.legend()
```
