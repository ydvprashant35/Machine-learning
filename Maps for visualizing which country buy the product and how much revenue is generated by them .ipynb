```python
df=df[(df['Country']!= 'United Kingdom')]
```

```python
df["Revenue"]=df["UnitPrice"]*df['Quantity']
df.head()
```

```python
# grouped_data=df.groupby('Country').agg({
#     "Revenue":"sum",
# }).reset.index()
grouped_data = df.groupby('Country').agg({
    'Revenue': 'sum',
}).reset_index()
grouped_data
```

```python
import folium
import pandas as pd
from geopy.geocoders import Nominatim

base_map = folium.Map(location=[20, 0], zoom_start=2)

# Initialize geolocator
geolocator = Nominatim(user_agent="geoapiExercises")

# Function to get latitude and longitude
def get_lat_lon(country):
    try:
        location = geolocator.geocode(country)
        return (location.latitude, location.longitude)
    except:
        return (None, None)

# Add markers for each country
for index, row in grouped_data.iterrows():
    country = row['Country']
    revenue = row['Revenue']
    lat, lon = get_lat_lon(country)
    
    if lat is not None and lon is not None:
        tooltip_content = f"{country}<br>Revenue: {revenue}"
        folium.Marker(
            location=[lat, lon],
             popup=country,  # Optionally, you can set popup to just country name
            tooltip=tooltip_content,
        ).add_to(base_map)

# Save the map to an HTML file
base_map.save("country_sales_map.html")
base_map
```
